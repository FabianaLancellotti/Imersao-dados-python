```python
import numpy as np

#Criao de um DataFrame de teste
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#Calcula a mdia salarial e substitui os nulos pela mdia, arredondando os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#Calcula a mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
#Criao de um DataFrame de teste
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28,27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()

df_temperaturas
```

```python
#Criao de um DataFrame de teste
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28,27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()

df_temperaturas
```

```python
#Criao de um DataFrame de teste
df_cidades = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'cidade': ["So Paulo", np.nan, "Curitiba", np.nan, "Belm"]
})

df_cidades['cidade_preenchida'] = df_cidades["cidade"].fillna("No informado")

df_cidades
```

```python
import plotly.express as px
```

```python
#Prompt IA: Crie um grfico de mdia salarial por senioridade em barras usando o plotly
df_media_salario = df_limpo.groupby('nivel_experiencia')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(df_media_salario,
    x='nivel_experiencia',
    y='usd',
    title='Salrio mdio anual por senioridade (USD)',
    labels={'nivel_experiencia': 'Nvel de Experincia', 'usd': 'Salrio Mdio Anual (USD)'})

fig.show()
```

```python
   #color='nivel_experiencia', # Para colorir cada barra por senioridade
   #color_discrete_sequence=px.colors.qualitative.Pastel # Escolha de uma paleta de cores

   #fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':df_media_salario['nivel_experiencia']})

   #Prompt IA: Crie um grfico de mdia salarial por senioridade em barras usando o plotly
df_media_salario = df_limpo.groupby('nivel_experiencia')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(df_media_salario,
    x='nivel_experiencia',
    y='usd',
    title='Salrio mdio anual por senioridade (USD)',
    labels={'nivel_experiencia': 'Nvel de Experincia', 'usd': 'Salrio Mdio Anual (USD)'},
    color='nivel_experiencia', # Para colorir cada barra por senioridade
    color_discrete_sequence=px.colors.qualitative.Pastel) # Escolha de uma paleta de cores

fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':df_media_salario['nivel_experiencia']})
fig.show()
```

```python
#Criao de grfico de pizza para tipo de trabalho

remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
                 values='quantidade',
                 title='Proporo dos Tipos de Trabalho')

fig.show()
```

```python
#Converter o grfico para rosca

remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
              values='quantidade',
              title='Proporo dos Tipos de Trabalho',
             hole= 0.5)

fig.show()
```

```python
#Para incluir a porcentagem + Label (nome das categorias) no grfico de rosca

remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
              values='quantidade',
              title='Proporo dos Tipos de Trabalho',
             hole= 0.5)

fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
#DESAFIO: Criar um grfico interativo com a mdia salarial por localizao_empresa para o cargo Data Scientist usando plotly. (Usado a IA para gerar esse grfico)

df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
df_media_salario_localizacao = df_data_scientist.groupby('localizacao_empresa')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    df_media_salario_localizacao,
    x='localizacao_empresa',
    y='usd',
    title='Mdia Salarial para Data Scientists por Pas (USD)',
    labels={'localizacao_empresa': 'Pas', 'usd': 'Salrio Mdio Anual (USD)'},
    color='localizacao_empresa',
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
